# Proyecto S12 DashBoard

## Introducción

Trabajo como analista de videos publicitarios en la agencia Sterling & Draper. Una parte clave de mi trabajo es detectar oportunidades de marketing a través del análisis de tendencias en YouTube: qué videos están destacando, en qué regiones y bajo qué categorías. Con frecuencia, colegas como Melanie y Ashok me solicitan reportes para responder preguntas como:

- ¿Qué categorías fueron tendencia la semana pasada?
- ¿Qué tan distribuidas estuvieron entre los distintos países?
- ¿Qué contenido fue especialmente popular en Estados Unidos?

Después de varias semanas generando estos análisis manualmente, decidí automatizar el proceso mediante la construcción de un **dashboard dinámico y consultable**.

Este dashboard se alimenta de una base de datos especializada (`youtube.trending_by_time`), la cual se actualiza diariamente con datos agregados sobre videos en tendencia, incluyendo fecha, región, categoría y número de videos destacados.

El objetivo del dashboard es permitir que los gerentes de planificación publicitaria puedan identificar rápidamente patrones de consumo y tomar decisiones informadas para sus campañas. Para lograr esto, el dashboard presenta visualizaciones clave y filtros interactivos que permiten explorar:

- El historial de tendencias por día y categoría (en valores absolutos y relativos).
- La distribución de tendencias por país (valores relativos).
- La relación entre categorías y países (valores absolutos, resaltados por intensidad).

Con esta solución, agilizo el acceso a insights clave sobre la popularidad de contenido, y facilito la planeación estratégica para futuras campañas publicitarias globales.
