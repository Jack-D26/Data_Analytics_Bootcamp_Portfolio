### 1. Use Prompt Templates

Adopt a consistent format when defining a task. For example:

[Objective]: Create modular data ingestion and cleaning script.
[Tools]: Python, Pandas.
[Expected Output]: Two Python scripts with inline documentation.
[Constraints]: Avoid hardcoding paths. Follow PEP8.

This helps AI better parse intent, especially in complex or multi-stage requests.


### 2. Explicit Output Expectations
Instead of just saying “write a README,” try:
“Generate a README that includes a table of contents, project overview, features, example usage, and technical stack.”
This narrows ambiguity and aligns with how pro prompt engineers use output shaping.


### 3. Apply Few-Shot Prompting for Repetitive Tasks

If you're developing lots of notebooks or functions, include a few examples of how you want scripts or docstrings to look.


#### Example docstring format:
"""
Function: load_data
Params: url (str)
Returns: pd.DataFrame
Description: Loads data from a remote CSV and returns a cleaned DataFrame.
"""


### 4. Metadata for Reusability
Consider embedding brief metadata tags in your prompt corpus when building a suite of tasks:

#### task_type: preprocessing
#### language: Python
#### goal: remove nulls and duplicates safely

This mirrors how dataset labels are used in LLM finetuning and makes versioning easier.


### 5. Evaluate with Checklists

. Adopt industry prompt review frameworks like:

. Does it specify input/output types?


. Does it define edge cases?


. Are formatting/behavioral constraints clear?


. Use this to self-grade or auto-grade your prompts over time.

## 🧰 Prompt Engineering Templates & Frameworks

This guide includes structured templates and best practices to help you craft precise, modular, and powerful prompts—ideal for use with Data_Partner or similar GPT-based tools.


### 📌 1. Task Definition Template

[Objective]:
Summarize the goal of this task in one sentence.

[Inputs]:
List any inputs needed (e.g., files, URLs, parameters).

[Tools]:
Mention libraries, platforms, or languages to use.

[Output Format]:
Describe the expected format and structure of the output (e.g., Markdown, Python, SQL).

[Constraints]:
List edge cases or rules (e.g., no hardcoded paths, must log errors, use PEP8).

[Success_Criteria]:
What does a successful output look like?


### 2. Example – Data Cleaning Script

[Objective]:
Create a data cleaning script for telecom call logs.

[Inputs]:
- `df_calls` DataFrame from telecom source

[Tools]:
- Python
- Pandas

[Output Format]:
A function named `clean_df_calls(df_calls)` with inline docstrings, returning the cleaned DataFrame.

[Constraints]:
- Remove rows with missing 'internal'
- Drop duplicates
- Convert date columns properly

[Success_Criteria]:
Function returns clean, ready-to-use DataFrame with correct dtypes.


### ⚙️ 3. Output Formatting Checklist

Use this checklist when defining what the model should return.

Does the output need to be runnable code?


Should it include inline documentation or comments?


Do you need Markdown formatting for GitHub?


Is an example input/output expected?


Should the response include links, citations, or references?

### 🔁 5. Iteration Feedback Template

Copy/paste this format to guide improvements over multiple prompts.

- The assistant correctly parsed the input.
- Output format was aligned with expectations.

[What Didn’t]:
- Missed a key step in date conversion.
- Docstring was vague.

[Next Request]:
Refine the prompt to emphasize type enforcement and docstring format.


6. Advanced Pattern: Chain of Thought Prompting

For complex pipelines or logic sequences:


Explain your thought process before writing code.

1. Identify types of missing data

2. Decide what can be filled vs. dropped

3. Apply transformations with minimal data loss

Then output Python code in one block.


### 🔄 7. Few-Shot Prompting (Pattern Recognition)

If you have consistent formatting preferences, define one or two examples:

Then prompt:
Apply the same column formatting logic to this new DataFrame.
